# 문제2 -15점

* 다음은 2022년 7월 26일 오후 2시 기준, 원화 환율입니다.
  (미국:1309.70$, 유럽연합:1339.30€, 중국:193.84¥, 영국:1579.96£)
* 환전하려는 금액을 입력 받고, 나라를 입력 받아서 환율을 **소수점 2자리**로 출력하세요. 
* Hint: 환전 금액과 나라 입력 시 input 함수를, 국가별 환율 계산에는 조건문을 사용하세요.

<답안 예시>

환전 금액(원화):8000

국가 선택(미국/유럽연합/중국/영국):미국

6.11


```python
usd = 1309.70 ; euro = 1339.30 ; cin = 193.84 ; eng = 1579.96

won = float(input('환전 금액(원화):'))
con = input('국가 선택(미국/유럽연합/중국/영국):')

if con == '미국' :
    print(round(won/usd,2))
elif con == '유럽연합':
    print(round(won / euro,2))
elif con == '중국':
    print(round(won / cin,2))
elif con == '영국':
    print(round(won / eng,2))
    
```

# 문제3 - 25점

### 문제 [3-1] - 3점

midwest.csv를 불러와 midwest에 저장하고, 위에서부터 10행까지 출력하시오.

```python
th = pd.read_csv('midwest.csv')
th.head(10)
```

### 문제 [3-2] - 7점

* popwhite는 해당 지역의 아시아인 인구, poptotal은 해당 지역의 전체 인구를 나타냅니다. 

* midwest 데이터에 '전체 인구 대비 아시아인 인구 백분율' 변수인 percent 열을 추가하세요.

```python
th['percent'] = th['popwhite'] / th['poptotal']
th.head()
```

### 문제 [3-3] - 3점

* percent 열의 값을 내림차순으로 정렬하세요.

```python
th['percent'].sort_values(ascending = False)
```

### 문제 [3-4] - 6점

* percent를 아래의 분류 기준에 따라 새로운 변수를 추가해주시고, 각 분류에 해당하는 지역이 몇 개인지 확인해보시오.

* large: 0.9% 이상 / middle: 0.4~0.9% 미만 / small: 0.4% 미만

```python
th['range'] = 'middle'
th.loc[th.percent>= 0.9 , 'range'] = 'large' 
th.loc[th.percent < 0.4 , 'range'] = 'small'  
th['range'].value_counts()
```

### 문제 [3-5] - 6점

* county가 'SCOTT'인 경우와 county가 'PIATT'인 경우의 poptotal 열의 평균을 각각 구해서 출력하세요.

* hint: query 함수와 mean 함수 사용

```python
print('SCOTT의 poptotal 평균', th.loc[th['county'] == 'SCOTT', 'poptotal'].mean())
print('PIATT의 poptotal 평균', th.loc[th['county'] == 'PIATT', 'poptotal'].mean())
```
